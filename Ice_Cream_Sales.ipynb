```python
import io
df2 = pd.read_csv(io.BytesIO(data_to_load['icecreamsales.csv']))
```

```python
df2.head()
```

# Total Profit Across the Year

Create pivot table of Total Profit by Month

```python
df_pv = df2.pivot_table(
    values=['Total Profit'],
    index=['Month'])

df_pv
```

Plot pivot table to line graph

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=df_pv)

ax.set(xlabel='Month', ylabel='Total Profit')
```

Create pivot table to split profit per month by flavor

```python
df_pvt = df2.pivot_table(
    values=['Vanilla', 'Chocolate', 'Strawberry', 'Neapolitan', 'Mango', 'Lime'],
    index=['Month'])
df_pvt
```

Plot new pivot table to line graph

```python
ax = sns.lineplot(data=df_pvt)

ax.set(xlabel='Month', ylabel='Total Profit')
```
